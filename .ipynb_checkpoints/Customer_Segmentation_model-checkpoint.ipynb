```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(12,8))
sc.dendrogram(sc.linkage(x,  method='single', metric='euclidean', optimal_ordering=False))
plt.show()
```

```python
al = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='single')
x['predict'] = al.fit_predict(x)
```

```python
x.head()
```

```python
x['predict'].value_counts()
```

```python
sns.pairplot(data=x, hue='predict')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.2,min_samples=4)
x['Pre'] = db.fit_predict(x[['Annual Income (k$)','Spending Score (1-100)']])
```

```python
x
```

```python
sns.scatterplot(x='Annual Income (k$)',y = 'Spending Score (1-100)', hue='Pre', data=x)
plt.show()
```

```python
x.drop(columns=['Pre','predict'], inplace=True)
```

```python
x
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(2,21):
    km = KMeans(n_clusters=i, init='k-means++')
    km.fit(x)
    wcss.append(km.inertia_)
```

```python
plt.figure(figsize=(10,8))
plt.plot([i for i in range(2,21)], wcss, marker='o')
plt.xticks([i for i in range(2,21)])
plt.xlabel('Cluser No')
plt.ylabel(' No of Value')
plt.grid(axis='x')
plt.show()
           
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(dataset,labels=km.labels_)
```

```python
sc = []
for i in range(2,21):
    km1 = KMeans(n_clusters=i, init='k-means++')
    km1.fit(x)
    sc.append(silhouette_score(x, labels=km1.labels_))
```

```python
plt.figure(figsize=(12,8))
plt.plot([i for i in range(2,21)], sc, marker='*')
plt.xticks([i for i in range(2,21)])
plt.xlabel("No of Cluster")
plt.ylabel('silhouette_score')
plt.show()
```

```python
km2 = KMeans(n_clusters=5, init='k-means++')
x['Prediction'] = km2.fit_predict(x[['Annual Income (k$)','Spending Score (1-100)']])
```

```python
x
```

```python
sns.scatterplot(x='Annual Income (k$)',y = 'Spending Score (1-100)', hue='Prediction',data=x)
plt.title("Customer Groups")
plt.show()
```
